## 02 | 대입연산자, 복합대입연산자

```python
1. 대입연산자
- =: 변수에 값을 할당할 때 사용하는 연산자
```

```python
2. 복합대입연산자
- 대입연산자와 함께 산술연산자를 결합하여 다양한 연산을 수행할 수 있다.
- 이러한 연산을 사용하면 더욱 간ㄴ단하게 표현할 수 있음
```

```python
- = 대입연산자
- +=: 
    A += 2 변수 A에 2를 더한 값을 저장 
    -> A = A + 2
        A = 4 -> A = 6
- -=: A -= 2  변수 A에 2를 뺀 값을 저장 
- *=: A *= 2  변수 A에 2를 곱한 값을 저장 
- **=: A **= 2  변수 A에 2를 거듭제곱한 값을 저장
- /=: A /= 2  변수 A에 2를 나눈 값을 저장
- //=: A //= 2  변수 A에 2를 나눈 몫을 저장
- %=: A %= 2  변수 A에 2로 나눈 나머지를 저장
```

## 03 | 관계연산자

```python
- 두 개의 값을 비교하여 그 결과 가 참인지 거짓인지 판별하는 연산자
- 비교연산자라고도 부름
```

```python
>: 크다
>=: 크거나 작다
<: 작다
<=: 작거나 같다
!=: 같지 않다
==: 같다
```

```python
num1 = 10
num2 = 20
print(num1 > num2) # 10 > 20 -> 거짓
print(num1 >= num2) # 크거나 같다
print(num1 < num2) # 10 < 20 -> 참
print(num1 <= num2) # 10 <= 20 -> 참
print(num1 != num2) # 10 != 20 -> 참
print(num1 == num2) # 10 == 20 -> 거짓
```

```python
num3 = 10
print(num3 == 10)
print(num3 == '10') 
```

```python
myage = int(input('본인의 나이를 입력해주세요: '))
if myage % 2 == 0:
    print('제 나이는 짝수입니다.')
else:
    print('제 나이는 홀수입니다.')
```

## 04 | 논리연산자

```python
01) 논리연산자란?
- 논리적인 연산을 수행하여 참인지 거짓인지를 판단하는 연산자
- 두 개 이상의 논리값을 조합하여 새로운 논리 값을 만들 수 있음
- 논리연산자의 종류
  - and: 연산자를 기준으로 왼쪽과 오른쪽 값이 모두 True일때만 True를 반환하고, 나머지는 False를 반환
  - or: 연산자를 기준으로 왼쪽과 오른쪽 값 중 하나라도 True이면 True를 반환하고, 둘 다 False일때만 False를 반환
  - not: 뒤에 따라오는 논리값이 True이면 False를 False이면 True를 반환
```

### 01) and: a와 b 모두 True면 True, 아니면 False

```python
print(True and True)
print(True and False)
print(False and True)
print(False and False)
```

### 02) or: a와 b 중 하나라도 True면 True, 아니면 False

```python
print(True or True)
print(True or False)
print(False or True)
print(False or False)
```

### 03) not: a가 True이면 False를, a가 False이면 True를 반환

```python
print(not "") # 문자열이 비어있으면 False를 반환이지만, not을 붙였기 때문에 True 반환
```

```python
print(not 'hello')
```

```python
print(not [])# 리스트가 비어있으면 False를 반환이지만, not을 붙였기 때문에 True 반환
```

```python
print(not [1, 2, 3])
```

```python
a = []
if not a: # a => False -> True
    print('a는 비어있습니다.')
else:
    print('a에 값이 있습니다.')
```

```python
a = [1, 2, 3]
if not a:  # a => True -> False
    print('a는 비어있습니다.')
else:
    print('a에 값이 있습니다.')
```

## 기타 | 삼항연산자

```python
- 조건문을 간단하게 작성할 수 있는 연산자
- 일반적으로 조건문 if-else문을 사용하여 작성하지만, 삼항연산자를 사용하면 한 줄로 조건문을 작성 가능
- 작성법: [참일때] if [조건문] else [거짓일때]
```

```python
a = 10
print("크다" if a > 5 else "작다") # a가 5보다 크면 '크다', 작으면 '작다'
```

## 연산자의 우선순위

```python
- 연산자를 두 개 이상 함께 사용할 때는 연산자의 순서가 정해져 있음
- 이 순서를 연산자의 우선순위라고 함
- 우선순위에 따라 연산이 수행되고 그 결과가 달라질 수 있기 때문에, 개발자에게 연산자의 우선순위는 매우 중요함
- 연산자의 우선순위를 이해하면 코드를 보다 효율적으로 작성할 수 있으며, 예상치 못한 결과를 방지할 수 있음
```

```python
- 우선순위
1. 괄호
2. 거듭제곱
3. 곱셈, 나눗셈, 나머지
4. 덧셈, 뺄셈
5. 비교연산자
6. not
7. and
8. or
9. (복합)대입연산자
```

```python
- 괄호
- 복잡한 연산을 수행할 때는 우선순위를 한눈에 파악하기 어려울 수 있음
  -> 따라서 먼저 처리해야 할 연산을 소괄호로 묶어서 우선순위를 가시적으로 표시하는 것이 좋음
- 괄호 안에 있는 연산은 최우선 순위로, 다른 연산자보다 먼저 처리됨
```

```python
- 연산자의 결합방향
  -> 연산자를 기준으로 결합 방향이 왼쪽에서 오른쪽으로 결합
```

```python
value = 80 - 30 + 10
print(value)
```

```python
value =  80 - 30 * 10
print(value)  # 곱하기 -> 빼기
```

```python
- 괄호를 사용해서 우선순위와 결합방향을 조절할 수 있음
```

```python
value = (80 - 30) * 10  # 괄호 -> 곱하기
print(value)
```

```python
value = 8 // 2 * (2 + 2)  # 괄호 -> 나누기(몫) -> 곱하기
print(value)
```

```python
1. () -> 4  # 8 // 2 * 4
2. // -> 4  # 4 * 4
3. * -> 16  # 16
```
