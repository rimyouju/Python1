## 01 | 생성자
- 생성자는 항상 __init__라는 이름을 가짐
- 첫 번째 매개변수는 항상 self
- 생성자는 인스턴스(객체) 속성(변수)을 초기화하는데 사용하고 유용함
- 생성자는 값을 반환하지 않으며, 반환문을 사용하면 안됨
- 객체가 생성될 때 자동으로 생성자인 init()이 호출됨
- => init(생성자)는 클래스를 통해 인스턴스를 생성할 때 무조건 생성되는 것

```python
# 생성자를 사용한 계산기
class FourCal():
    def __init__(self, first, second):
        self.first = first
        self.second = second
    def add(self):
        result = self.first + self.second
        return result
a = FourCal(4, 2)
a.add()
```

## 02 | 상속이란?
- 이미 만들어진 클래스에 다른 기능들을 추가하고 싶을 때 사용 가능
- 즉, 부모 클래스가 가진 기능들을 모두 물려 받을 수 있음

```python
01) 상속 사용하기
```

```python
[작성법]
class 부모클래스명:
    사용할 메서드1:
    사용할 메서드2:
    ...
class 자식클래스명(부모클래스명):
    사용할 메서드3:
    사용할 메서드4:
    ...
```

```python
class Person():
    def __init__ (self, name, age):
        self.name = name
        self.age = age
    def greet(self):
        print(f"Hello my name is {self.name}")
class Student(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age) # super(). => 부모클래스에 접근하겠다
        self.subject = subject
    def study(self):
        print(f"{self.name} 학생은 {self.subject} 공부를 한다")
s = Student("Alice", 20, "Math")
s.greet()
s.study()
print(s.name)
print(s.age)
print(s.subject)
```

## 03 | 다중상속이란?
- 다중 상속은 말 그대로 여러 개의 클래스를 상속하는 것을 말함
- 사용방법은 상속할 대상을 , 를 사용하여 나열

```python
01) 다중상속 사용
```

```python
[작성법]
class 기반클래스이름1:
    ...
class 기반클래스이름2:
    ...
class 파생클래스이름(기반클래스이름1, 기반클래스이름2):
    ...
```

```python
class Animal:
    def bark(self):
        print("bark")
class Mammal:
    def feed(self):
        print("feed")
class Cat(Animal, Mammal):
    def name(self):
        print("Cat")
cat1 = Cat()
cat1.bark()
cat1.feed()
cat1.name()
```

```python
class Person:
    def greeting(self):
        print("안녕하세요")
class Worker:
    def greeting2(self):
        print("저는 일하는 중입니다")
class Student():
    def study(self):
        print("저는 공부하는 중입니다")
class Intern(Person, Worker, Student):
    def greeting3(self):
        super().greeting()
        print("저는 인턴입니다")
jane = Intern()
jane.greeting3()
jane.study()
jane.greeting2()
```

## 04 | 오버라이팅
- 무시하다, 우선하다
- 부모 클래스의 메서드를 무시하고 새로운 메서드를 만든다
- 부모 클래스의 메서드를 자식 클래스에서 재정의 하는 것

```python
class Animal:
    def sound(self):
        print("오버라이딩 전")
        print("멍멍")
class Dog(Animal):
    def sound(self):
        print("오버라이딩 후")  # 기존 내용이 유지되면서 재정의 함
        print("멍멍멍멍멍")
class Cat(Animal):
    def sound(self):
        print("야옹") 
d = Animal()
d.sound()
d2 = Dog()
d2.sound()
c = Cat()
c.sound()
```

#### 연습문제 | 
- 클래스 C와 클래스 Car을 생성
- 클래스 Car는 클래스 C를 상속받음
- 클래스 C에는 생성자를 통해 데이터를 저장하는 기능 생성
- 클래스 Car는 info_print 메서드를 생성하고 출력 기능을 만든다

```python
[출력내용]
바퀴수 4
가격 10000
# 객체 생성 및 출력
car = Car(4, 10000)
car.info_print()
```

```python
class C:
    def __init__(self, wheel, price):
        self.wheel = wheel
        self.price = price
class Car(C):
    def info_print(self):
        print(f"바퀴수 {self.wheel}")
        print(f"가격 {self.price}")
car = Car(4, 10000)
car.info_print()
```

#### 연습문제
- 상속의 개념을 활용하여, 기본 인적 사항(이름, 나이)를 가진 Person 클래스를 만들고, 이를 상속받는 Student 클래스를 작성하여 학생 번호까지 출력

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
class Student(Person):
    def info(self, name, age, num):
        super().__init__(name, age)
        self.num = num
    def info_print(self):
        print(f"이름: {self.name}")
        print(f"나이: {self.age}")
        print(f"학번: {self.num}")
student1 = Student("youju", 18, 1234)
student1.info_print()
```
