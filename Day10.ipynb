```python
계산기를 만들고 기능 상속받기
```

```python
class FourCal:
    def __init__(self, first, second):
        self.first = first
        self.second = second
    # 덧셈
    def add(self):
        result = self.first + self.second
        return result
    # 뺄셈
    def sub(self):
        result = self.first - self.second
        return result
    # 곱셈
    def mul(self):
        result = self.first * self.second
        return result
    # 나눗셈
    def div(self):
        result = self.first / self.second
        return result

# 상속받기
class FourCal2(FourCal):
    def pow(self):
        result = self.first ** self.second
        return result
num = FourCal2(3, 4)
print(num.add())
```

#### 연습문제 |

```python
[출력결과]
안녕하세요 제 이름은 김민수이고 나이는 30살 입니다.
안녕하세요 제 이름은 이영희이고 나이는 25살 입니다.
안녕하세요 제 이름은 박철수이고 나이는 40살 입니다.
```

```python
class Person():
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def introduce(self):
        return f"안녕하세요 제 이름은 {self.name}이고 나이는 {self.age}살 입니다."
        
# Person 클래스를 사용해 객체 생성 후 리스트에 저장
people = [] # people = [객체1, 객체2, 객체3]
people.append(Person("김민수", 30))
people.append(Person("이영희", 25))
people.append(Person("박철수", 40))

# 리스트에 저장된 Person 객체들의 정보를 출력
for p in people:
    print(p.introduce())
```

```python
print(people)
```

#### 연습문제 | 연락처 저장 및 검색

```python
- class Contact 클래스 생성
- 연락처 3개를 저장할 빈 리스트 생성
- for문 사용
    name, phone입력받기
    빈 리스트에 추가(객체생성)
- 이름으로 검색
    - 검색하고자 하는 이름 입력받기
    - found = False
    - for 리스트에 값이 있는 동안:
    - if 비교하기(==)
        print
        found = True
        break
- if not found:
 print("연락처에 없습니다.")
```

```python
[출력결과]
이름: 수지
전화번호: 010-0001-1111
이름: 지아
전화번호: 010-0002-2222
이름: 유나
전화번호: 010-0003-3333

검색할 이름: 수지
수지의 전화번호는 010-0001-1111입니다
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def show_info(self):
        print(f"이름: {self.name},  점수: {self.score}")
```

```python
contacts = []
for i in range(3):
    print("연락처를 입력하세요")
    name = input("이름: ")
    phone = input("점수: ")
    c = Contact(name, phone)
    #리스트에 추가
    contacts.append(c)
```

```python
print(contacts)
```

```python
for n in contacts:
    print(n)
    print(n.name)
    print(n.phone)
```

```python
name = input("검색 할 이름:")
found = False
for n in contacts:
    if name == n.name:
        print(f"{n.name}의 전화번호는 {n.phone}입니다.")
        found = True
        break
    if not found: # not False -> True
        print("연락처에 없습니다")
```

#### 연습문제 | 최고 점수 학생 찾기

```python
- class Student 생성
  속성: name, score
- 학생 3명의 이름, 점수 입력받기
- 빈리스트(students)에 저장
- 최고 점수를 찾기 위한 값 비교
- 최고 점수 학생 출력
```

```python
[입력]
이름: 수지
점수: 95
이름: 지아
점수: 88
이름: 유나
점수: 100

[출력 결과]
최고 점수 학생: 유나(100점)
```

```python
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score
```

```python
students = []
for i in range(3):
    name = input("이름: ")
    score = int(input("점수: "))
    s = Student(name, score)
    students.append(s)
```

```python
print(students)
```

```python
top = students[0]
print(top.score)
for s in students[1:]:
    if s.score >= top.score:
        print(f"최고 점수 학생: {s.name}({s.score}점)")
        break
```
