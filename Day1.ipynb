```python
### 01) 문자열 인덱스
- 문자열은 문자들의 *순서있는 나열*이기 때문에, 각 문자는 번호(인덱스)를 가지고 있음
```

```python
slice1 = '파이썬 수업 중입니다'
print(slice1)

# 인덱스 번호에는 공백도 포함!!
```

### 02) 슬라이싱 연산

```python
[시작:끝:간격]
[이상:미만]
- 시작: 시작 인덱스, 생략하면 처음부터 추출
- 끝: 종료 인덱스, 생략하면 끝까지 추출(이거 직전까지 슬라이싱 해줘)
- 간격: 인덱스의 증감값, 생략하면 1씩 변화
```

```python
파이썬
0 1 2
```

```python
slice3 = "안녕하세요여기는파이썬교실입니다"
print(slice3[0:4])
print(slice3[1:7])
```

```python
print(slice3[1:7:2])
```

```python
slice4 = "파 이 썬 수 업 중 입 니 다"
print(slice4[::2])
```

```python
print(slice5[2:])
```

## 추가 06 | 문자열 포매팅

```python
- 문자열 안에 변수나 값을 넣는 방법
```

### 01) format() 메서드

```python
- format() 메서드는 문자열의 중괄호 () 안에 특정 값을 삽입하는 데 사용
```

```python
print('My name is {}'.format('james'))
```

```python
print('My name is {0}. I\'m {1}years old.'.format('james', 25))
```

```python
print('My name is {name}. I\'m {age} years old.'.format(name = 'james', age = 25))
```

### 02) f - 스트링

```python
- 문자열 앞에 f 또는 F를 붙이고, {}중괄호 안에 변수 또는 표현식을 넣어 값을 삽입하는 방식
```

```python
1. {} 중괄호 안에 직접 변수를 넣는 방법
```

```python
s = 'coffee'
n = s
result = f'저는 {s}를 좋아합니다. 하루 {n}잔 마셔요.'
print(result)
```

```python
2. 간단한 표현식 사용하기
```

```python
age = 25
print(f"내년엔 {age+1}살이 됩니다")
```
