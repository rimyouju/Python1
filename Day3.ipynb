## 03 | 반복문

### 01) while문

```python
- 특정 조건을 만족하는 동안 반복해서 수행
- 반복의 횟수가 사전에 정해지지 않았거나 특정조건이 만족될 때까지 반복해야하는 경우
- 반복문 실행이 종료되면 다시 조건식으로 이동, 조건식의 결과가 False이면 더이상 반복하지 않는다.

** 주의사항 **
    - 무한루프에 빠지지 않도록 주의!
    - 루프를 종료할 수 있는 조건을 명확히 하고, break문을 적절히 사용하여 필요할 때 루프를 빠져나올 수 있게 해야함
```

### while문의 구조

```python
while 조건식:  # 조건이 False가 될 때까지 반복
    반복실행문
```

```python
while True:
    print('hello')  # 무한루프에 빠짐
```

### 연습문제 03 | 입력받은 수까지 합 구하기

```python
- 사용자로부터 자연수n을 입력받아, 1부터 n까지의 합을 while문으로 구하기
```

```python
n = int(input('자연수를 입력하세요: ')) # 입력
num = n  # num에다 n 값을 할당
total = 0
while 0 < n:
    total += n 
    n -= 1
print(f"1부터 {num}까지의 합: ", total)
```

### range(시작, 끝, 증감식)

```python
- 괄호 안의 값이 하나인 경우 range(끝직전)값만 인식하게 됨
- 시작(start): 시작부분, 생략하면 0부터 시작
- 끝(stop): 끝부분, 끝은 포함하지 않고 직전까지만
- 증감식(step): 얼마씩 증가(또는 감소)할지, 생략하면 기본은 +1
```

```python
[작성법]
1. 끝직전 값만 입력
for 임시변수 in range(끝직전 값):
    반복할 코드
2. 시작, 끝직전 값
for 임시변수 in range(시작값, 끝직전 값):
    반복할 코드
3. 증감값 추가
for 임시변수 in range(시작값, 끝직전 값, 증감값):
    반복할 코드
```

```python
#끝직전값
for i in  range(10):
    print(i)
```

```python
# 시작값, 끝직전값
for i in range(3, 10):
    print(i)
```

```python
# 시작값, 끝직전값, 증감값
for i in range(2, 10, 2):
    print(i)
```

```python
# 시작값, 끝직전값, 증감값
#20씩 끊어서 출력
for k in range(0, 100, 20):
    print(k, end = ' ')  # new line대신 공백으로 나누어 출력
print('99까지이므로 끝!')
```

## 04 | 분기문

```python
1. break
2. continue
3. pass
```

```python
01) continue구문
- 아래를 이어서 실행하지 않고 바로 조건 검사문으로 넘어감
- 해당 반복은 종료하고 다음 반복을 실행
- for문과 while문 안에서 if문을 통해 조건 검사
```

```python
# for문과 continue
for i in range(1, 10):  # 1 ~ 9
    if i % 2 == 0:  # 만약에 짝수이면 continue문을 만남 -> 다음 조건 검사문으로 넘어가라
        continue
    print(i)
```

```python
# while문과 comyinue
i = 0
while i < 5:  # i가 5보다 작으면 print
    i += 1  # 그리고 i = i + 1
    if i == 2:  # 만약 i가 2이면 continue문을 만남 -> 다음 조건 검사문으로 넘어가라
        continue
    print(i)
```

```python
02) break문
- 반복문을 중단하고 싶을 때 사용하는 것
- for문과 while문 안에서 if문을 통해 조건 검사
```

```python
# for문과 break
for i in range(1, 6):
    if i == 3:
        break
    print(i)
```

```python
# while문과 break
num1 = 0
while True:  # 원래는 무한루프에 빠지게 됨
    print(num1)
    break  # break를 만나는 순간 코드 종료
```

```python
03) pass
- 반복문, 함수, if문에서 비워두고 싶을 때 사용
- 코드 블록이 비어있어도 문법적으로 오류가 나지 않게하기 위해 사용
- 특정 조건을 처리해야 하지만 아직 구현되지 않았거나 처리할 필요가 없을 때 사용
```

```python
# for, if문과 pass
for i in range(5):
    if i == 3:
        pass  # 수행할 것을 정하지 않아도 코드 수행 가능 (pass가 없으면 에러 발생)
    print(i)
```

### 연습 01 | 1부터 10까지 숫자 누적합을 출력

```python
total = 0
for i in range(1, 11): # 1-10까지
    #print(i)
    total += i 
print(total)
```

### 연습 02 | 1부터 100까지 숫자 중에서 짝수 누적합 출력

```python
total = 0
for i in range(1, 101):  # 1-100까지
    if i % 2 != 0:  
        continue   # i가 짝수가 아니면 다시 조건반복문으로
    total += i  # i가 짝수이면 +
print(total)
```

### 연습 03 | 1부터 9까지의 숫자 중 3의 배수를 제외한 숫자들의 합

```python
total = 0
for i in range(1, 10):  # 1-9
    if i % 3 == 0:  
        continue  # i가 3의 배수이면 다시 조건반복문으로
    total += i  # i가 3의 배수가 아니라면 +
print(total)
```

### 연습 04 | 1부터 50까지의 숫자 중, 3의 배수이거나 5의 배수인 숫자의 개수 출력

```python
count = 0
for i in range(1, 51):  # 1-50
    if i % 3 == 0 or i % 5 == 0:  # i가 3이나 5의 배수라면 count
        count += 1
print(count)
```

```python
number = int(input("숫자를 입력하세요: "))
count = 0 # 반복 횟수를 설정하기 위한 변수
# while문을 사용하여 number만큼 반복
while count < number:
    print(number, end = '')
    count += 1
```

### 과제 02 |
- 입력받은 정수를 계속 더해나가다가, 음수가 입력되면 중단하고 그 전까지 계산한 값을 출력

```python
[출력결과]
입력되는 숫자를 모두 더하는 프로그램입니다. -1을 입력하면 종료됩니다
숫자를 입력하세요: 1
숫자를 입력하세요: 2
숫자를 입력하세요: 3
숫자를 입력하세요: -1
6
```

```python
print('입력되는 숫자를 모두 더하는 프로그램입니다. -1을 입력하면 종료됩니다.')
total = 0
while True:
    num = int(input('숫자를 입력하세요: '))
    if num != -1:
        total += num
    elif num == -1:
        break
print(total)
```
