```python
## 02 | 재귀호출
01) 재귀호출이란?
- 자기 자신을 호출하는 함수
- 반복적/ 규칙적 문제를 작은 하위 문제로 나누어 해결할 때 사용

02) 특징
- 종료 조건이 반드시 필요
- 종료 조건이 없으면 무한 호출
- 문제를 하위 문제로 분할
- 큰 문제 -> 작은 문제 -> 더 작은 문제 -> 종료 조건
- 스택 구조로 동작
  -> 함수 호출이 쌓이고, 종료 조건에 도달하면 하나씩 반환
```

```python
[재귀호출의 기본 구조]
def 함수이름(매개변수):
    # 재귀호출을 멈추기 위한 조건
    if 종료조건:
        return 결과
    수행할 문장1
    수행할 문장2
    ...
    함수이름(인수) # 자기 자신을 함수에서 호출

# 함수 호출
객체이름 = 함수이름(인수)
```

```python
def recursive(n):
    if n == 0:
        return 0
    else:
        recursive(n-1)
        print(n)
a = recursive(5)
print("최종반환값:", a)

# 스택 구조라서 하나하나 값이 쌓임 -> 선입후출
# 5까지 다 출력됨 -> 이제 더이상 출력될 값이 남아있지 않음 -> 최종반환값 = none
```

```python
### 팩토리얼이란?
- 자연수 n에 대해 1부터 n까지 모든 자연수를 곱한 값
- n! = n * (n - 1) * (n - 2) * ... * 2 * 1
- 5! = 5 * 4 * 3 * 2 * 1
- 1! = 1
- 0! = 1 (정의에 의해)
```

```python
- 재귀호출을 사용하영 팩토리얼 구현
01. 설계
[종료 조건 정의]
- factorial(1) -> 1 반환
- 재귀호출을 멈추는 역할
[재귀호출 구조]
- n-1을 다음 단계로 전달하여 문제를 점점 줄임
```

```python
# 함수 설계
def 함수이름(n):
    # 1. 종료 조건
    if 입력된 값이 1인 경우 == 1:
        1을 반환
    # 2. 재귀호출
    else:
        return n * 함수호출(n - 1)
# 함수 호출
객체 = 함수이름(n)
```

```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n-1)
result = factorial(4)
print(result)
```

## 05 | 사칙연산 계산기 만들기

```python
1. 계산기 설계
- 사칙연산 가능
- 속성 2개: 두 개의 값을 입력받는 기능 만들기
- 메서드 4개: 덧셈, 뺄셈, 곱셈, 나눗셈
```

```python
- 클래스 정의 및 덧셈 기능 구현
```

```python
class FourCal():
    def setdata(self, first, second):
        self.first = first
        self.second = second
    def add(self):
        result = self.first + self.second
        return result
    def minus(self):
        result = self.first - self.second
        return result
    def multiply(self):
        result = self.first * self.second
        return result
    def division(self):
        result = self.first / self.second
        return result
a = FourCal()
a.setdata(4, 2)
print("------- a의 연산 결과 -------")
print("덧셈 결과:", a.add())
print("뺄셈 결과:", a.minus())
print("곱셈 결과:", a.multiply())
print("나눗셈 결과:", a.division())
```

```python
클래스 안에서 메서드 이용하기
```

```python
class FourCal():
    def setdata(self, first, second):
        self.first = first
        self.second = second
    def add(self):
        result = self.first + self.second
        return result
    def minus(self):
        result = self.first - self.second
        return result
    def multiply(self):
        result = self.first * self.second
        return result
    def division(self):
        result = self.first / self.second
        return result
    def print(self):
        print(f"덧셈 결과: {self.add()}")
        print(f"뺄셈 결과: {self.minus()}")
        print(f"곱셈 결과: {self.multiply()}")
        print(f"나눗셈 결과: {self.division()}")
a = FourCal()
a.setdata(4, 2)
a.print()
```

```python
class Computer():
    def set_spec(self, CPU, RAM, SSD):
        self.CPU = CPU
        self.RAM = RAM
        self.SSD = SSD
    def hardware_info(self):
        print("CPU =", self.CPU)
        print("RAM =", self.RAM)
        print("SSD =", self.SSD)
desktop = Computer()
desktop.set_spec("i7", "16GB", "512GB")
desktop.hardware_info()
```

- setdata() 메서드를 정의하여 학생의 이름, 나이, 점수를 설정
- is_passed() 메서드를 추가하여 점수가 60 이상이면 Passed를, 그렇지 않으면 Failed를 출력

```python
class Student():
    def setdata(self, name, age, score):
        self.name = name
        self.age = age
        self.score = score
    def is_passed(self):
        if self.score >= 60:
            print(self.name, "Passed")
        else:
            print(self.name, "Failed")
student1 = Student()
student1.setdata("john", 20, 75)
student1.is_passed()
```
