### 06) 튜플의 슬라이싱
슬라이싱
- [시작: 끝: 간격] -> [start: stop: step]
- [이상: 미만]
- start: 시작 인덱스; 생략하면 처음부터 추출
- stop: 종료 인덱스; 생략하면 끝까지 추출
- step: 인덱스의 증감값; 생략하면 1씩 변화

```python
p = (1, 2, 3, 4, 5, 6, 7)
print(p[1:4])
```

```python
p = (1, 2, 3, 4, 5, 6, 7)
print(p[:2])
```

```python
p = (1, 2, 3, 4, 5, 6, 7)
print(p[1::2])
```

```python
** step이 음수인 경우
- step 방향에 따라 해석이 달라짐
- step이 양수 -> start를 생략할 경우 0부터 시작, stop을 생략할 경우 끝까지
- step이 음수 -> start를 생략할 경우 마지막부터, stop을 생략할 경우 0번 인덱스까지
```

```python
p = (1, 2, 3, 4, 5, 6, 7)
print(p[::-1])
```

```python
# start = 4 -> 인덱스 4에서 시작
# stop = 처음(인덱스 0)까지
# step = -1 -> 뒤로(거꾸로) 한 칸씩 이동
p = (1, 2, 3, 4, 5, 6, 7)
print(p[4::-1])
```

### 06) 딕셔너리 다루기 기타

```python
- keys(): 딕셔너리의 모든 키를 반환
- values(): 딕셔너리의 모든 값을 반환
- items: 딕셔너리의 모든 키 - 값 쌍을 튜플의 리스트로 반환
- gets(): 특정 키에 대한 값을 반환, 키가 없으면 기본값을 반환
```

```python
1. key 값들만 추출
```

```python
nations = {'Korea': '한국', 'Japan': '일본', 'China': '중국'}
for i in nations:
    print(i)
```

```python
nations = {'Korea': '한국', 'Japan': '일본', 'China': '중국'}
for i in nations.keys():
    print(i)
```

```python
2. value들만 추출
```

```python
nations = {'Korea': '한국', 'Japan': '일본', 'China': '중국'}
for i in nations.values():
    print(i)
```

```python
3. items() | key와 value 동시 접근
- 키와 값을 순서대로 동시에 반환
```

```python
nations = {'Korea': '한국', 'Japan': '일본', 'China': '중국'}
for key, value in nations.items():
    print(f"키: {key}, 밸류: {value}")
```

```python
4. get() | 특정 키에 대한 값을 반환
- 특정 키의 값을 가져옴
- 키가 없으면 기본값을 반환
```

```python
[작성법]
- 딕셔너리명.get(key, 기본값)
- 딕셔너리명.get(key) -> 기본 반환값은 None
```

```python
student = {
    'name': 'Jone', 
    'age': 22, 
    'major': 'computer science'
}
# 키가 존재하는 경우
name = student.get('name')
print(name)
# 키가 존재하지 않는 경우 -> none반환
graduation_year = student.get('id')
print(graduation_year)
# 키가 존재하지 않는 경우, 기본값 사용
n = student.get('id', 'Not specified')
print(n)
```

```python
3. remove와 discard의 차이점
- 에러 발생 유무
- remove는 삭제할 대상이 없으면 에러 발생
- discard는 에러가 발생하지 않음
```

```python
4. 각각의 사용 이유
- remove()
-> 원소가 반드시 집합 안에 있어야 하고, 없으면 오류로 알려주길 원할 때 사용
-> 데이터 무결성(정확성)이 중요한 상황에 사용

- discard()
-> 원소가 집합 안에 있을 수도, 없을 수도 있을 때 안전하게 삭제하고 싶을 때 사용
-> 불필요한 오류를 피하면서 코드 단순화 가능
```

### 05) 집합연산

```python
- 교집합: A에도 속하고 B에도 속하는 원소들을 골라 놓은 집합
- 합집합: A에 속하거나 B에 속하는 원소들로 이루어진 집합
- 차집합: 한 집합에서 다른 집합 요소를 뺀 새로운 집합을 생성하는 집합연산
         A - B
         B - A
```

```python
setA = {10, 20, 30, 40, 50, 60}
setB = {50, 60, 70, 80, 90, 100}
# 교집합
print(setA & setB)
```

```python
# 합집합
print(setA | setB)
```

```python
# 차집합
print(setA - setB)
print(setB - setA)
```

### 06) 집합연산 메서드

```python
- intersection() | 교집합
```

```python
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5, 6, 7}
set3 = set1.intersection(set2)
print(set3)
```

```python
- union() | 합집합
```

```python
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5, 6, 7}
set3 = set1.union(set2)
print(set3)
```

```python
- difference() | 차집합
```

```python
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5, 6, 7}
set3 = set1.difference(set2)
print(set3)
```

#### 과제 01 | 각 학생들의 성적이 주어졌을 때 평균 계산

```python
scored = {'Alice': [85, 90, 92], 
          'Bob': [78, 81, 75], 
          'Charlie': [92, 95, 77]
 }
```

```python
[출력결과]
Alice의 점수들은 [85, 90, 92]입니다.
Bob의 점수들은 [78, 81, 75]입니다.
Charlie의 점수들은 [92, 95, 77]입니다.

{'Alice': 89.0, 'Bob': 85.17, 'Charlie': 86.11}
```

```python
scored = {'Alice': [85, 90, 92], 
          'Bob': [78, 81, 75], 
          'Charlie': [92, 95, 77]
 }
for key, value in scored.items():
    print(f"{key}의 점수들은 {value}입니다.")

avg = {}
for student, scores in scored.items():
    total = 0
    count = 0
    for i in scores:
        total += i
        count += 1
        score = total / count
    avg[student] = round(score, 3)
print(avg)
```

```python
scored = {'Alice': [85, 90, 92], 
          'Bob': [78, 81, 75], 
          'Charlie': [92, 95, 77]
 }
for key, value in scored.items():
    print(f"{key}의 점수들은 {value}입니다.")

avg = {}
for student, scores in scored.items():
    
    avg[student] = round(sum(scores)/len(scores), 3)
print(avg)
```

```python
round() | 반올림 하는 함수
[작성법]
변수명 = round(number, digits)
number -> 반올림 할 숫자
digits -> 반올림 할 소숫점 자릿수, 생략하면 정수로 반올림해줌
round(number, 2) -> number을 소숫점 두 자리까지 반올림 해
```
