## 02 | 문자열 메서드(문자열 조작함수)

```python
- format(): 문자열의 값을 삽입할 때 사용
- count(): 특정 문자열이 등장하는 횟수 반환
- find(): 문자열에서 처음 등장하는 위치 반환, 없다면 -1 반환
- index(): find와 비슷하지만, 없을 경우 ValueError 발생
- join(): 문자열 사이에 값을 넣어 interablef(반복 가능 객체)의 요소들을 연결
- split(): 특정 구분자를 기준으로 문자열 나눔
- replace(): 
- 
```

```python
format(): 문자열의 값을 삽입할 때 사용
- 정렬하는 방법
```

```python
name = 'Alice'
age = 25
format_s = "Name: {}, Age: {}".format(name, age)
print(format_s)
```

```python
- format(): 정렬
왼쪽 정렬: <
오른쪽 정렬: >
가운데 정렬: ^

- 10의 자리의 필드폭(자료를 출력하는데 사용할 전체 칸) 의미
[작성법]
변수명 = {포멧사양}.fomat('출력할 문자')
[예] {:<10}.format('left')
[설명] {:포멧 사양 시작< 외ㄴ쬬ㅗㄱ정려ㅑㄹ 10의자리촉}
```

```python
# 왼쪽 정렬
left = "{:<10}".format("left")
print(f"'{left}'")
```

```python
# 오른쪽 정렬
right = "{:>10}".format("right")
print(f"'{right}'")
```

```python
center = "{:^10}".format("center")
print(f"'{center}'")
```

```python
count(): 특정 요소의 개수를 세는데 사용
[작성법]
결과값을 담을 변수명 = 문자열 변수.count("찾을 문자열 또는 문자")
```

```python
text = "hello world. hello python"
count_hello = text.count('hello')
count_o = text.count('o')
print('hello', count_hello)
print('o', count_o)
```

```python
find(): 특정 문자열은 찾고자 할 때 사용
[작성법]
변수명 = 문자열변수명.find('찾을글자')
변수명 = 문자열변수명.find('찾을글자', 기준인덱스)
변수명 = 문자열변수명.find('찾을단어')
* 찾는 문자열이 없는 경우 -1 반환
```

```python
# 문자 찾기
text = "hello world"
index = text.find('l') # 가장 첫 번째 인덱스 반환
print(index)
```

```python
text = "hello world"
index = text.find('l', 5) # 가장 첫 번째 인덱스 반환
print(index)
```

```python
text = "hello world"
index = text.find('world') #대소문자도 구분함
print(index) 
```

```python
- 대소문자 구분함
```

```python
index(): 특정 문자를 찾을 때 사용
- find와 차이점
-> find는 찾고자 하는 문자열이 없는 경우 -1을 반환하지만, .index는 오류가 발생함
```

```python
text = "hello world"
index_h = text.index('h')
index_o = text.index('o', 5)
index_world = text.index('world')
print(index_h)
print(index_o)
print(index_world)
```

```python
join(): 여러 문자열을 하나의 문자열로 결합할 때 사용
[작성법]
변수명 = "구분자".join(변경하고 싶은 변수명)

* 구분자는 다양하게 삽입 가능
" ": 공백
",": 쉼표
"/": 슬래시
"-": 하이픈
"\n": 새 줄
```

```python
# 리스트의 문자열 요소들을 공백으로 결합
w = ["Hello", "world", "Python", "is", "awesome"]
sentence = " ".join(w)
print(sentence)
```

```python
# 공백 없이 조인
w = ["H", "e", "l", "l", "o"]
sentence = "".join(w)
print(sentence)
```

```python
# 쉼표를 사용하여 결합
w = ["Hello", "world", "Python", "is", "awesome"]
a = ",".join(w)
print(a)
```

```python
w = ["Hello", "world", "Python", "is", "awesome"]
a = "-".join(w)
print(a)
```

```python
c = "hello"
d = "-".join(c)
print(d)
```

```python
str1 = "hello my name is. ."
str2 = "\n".join(str1)
print(str2)
```

```python
split(): 문자열 분리
```

```python
- 문자열을 특정 구분자를 기준으로 잘라서 리스트로 변환하는 매서드
```

```python
text = "hello world, how are you"
# 기본 공백 문자를 기준으로 문자열 분리
words = text.split()
print(words)
```

```python
str1 = '사과-참외-딸기-메론-수박'
arr1 = str1.split('-')
print(arr1)
```

```python
# 분리 개수 제한 -> 최대 두 번 분리
text = "one, two, three, four, five"
s_t = text.split(", ", maxsplit = 2)
print(s_t)
```

```python
replace(): 문자열 바꾸기
```

```python
- 문자열 내의 특정 부분 문자열을 다른 문자열로 바꾸는데 사용
- 원본 문자열을 변경하지 않고, 변경된 새로운 문자열을 반환
```

```python
text = "Hello! World!"
result = text.replace("World", "Python")
print(result)
```

```python
result = text.replace("o", "ee")
print(result)
```

```python
result = text.replace("l", "t", 2) # 2번째 l까지만 t로 변경하겠다
print(result)
```

```python
불필요한 문자 제거: strip(), lstrip(), rstrip()
```

```python
- strip()
- 앞뒤 공백 삭제, 또는 특정 문자 제거
```

```python
text = "    Hello, world !!    "
result = text.strip()
print(text)
print(result)
```

```python
- lstrip()
- 문자열의 왼쪽 끝에서 공백이나 지정한 문자을 제거
```

```python
# 왼쪽 공백 삭제
text = "    Hello, world !!    "
text2 = text.lstrip()
print(f"'{text2}'")
```

```python
# 왼쪽 특정 문자 제거
text = "xxxhello worldxxx"
ls = text.lstrip('x')
print(f"'{ls}'")
```

```python
- rstrip()
- 문자열의 오른쪽 끝에서 공백이나 지정한 문자를 제거
```

```python
text = "    Hello World   "
text2 = text.rstrip()
print(F"'{text2}'")
```

```python
# 오른쪽 특정 문자 제거
text = "xxxhello worldxxx"
rs = text.rstrip('x')
print(f"'{rs}'")
```
